```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
import matplotlib.dates as mdates
import matplotlib.ticker as mtick

plt.plot()
```

```python
APR4 = pd.read_csv("t_hplus_APR4_q09.txt", sep=" ", header = None)
APR4.columns=['time', 'strain']
plt.figure()
ax = plt.gca()

plt.rcParams['text.usetex'] = True
APR4.plot(ax=ax,
          x='time', y='strain',
          kind='line',
          linewidth = 0.8,
          color='red',
          xlabel='time t [sec.]',
          ylabel='strain $h_{+}$',
          legend=False)
ax.grid(linestyle='dotted', alpha=0.75)
plt.ylim(bottom=-5e-18)#, top=5e-18)
ax.annotate('inspiral', (0.003, -4.2e-18))
ax.annotate('merger', (0.013, -4.2e-18))
ax.annotate('post-merger', (0.035, -4.2e-18))
x1, y1 = [0, 0.014], [-3.5e-18,-3.5e-18]
x2, y2 = [0.0145, 0.0175], [-3.5e-18,-3.5e-18]
x3, y3 = [0.018, 0.06], [-3.5e-18,-3.5e-18]
plt.plot(x1, y1, x2, y2, x3, y3, marker = '|', linestyle='dashed')

plt.savefig('APR4.pdf', dpi=1000)
```
