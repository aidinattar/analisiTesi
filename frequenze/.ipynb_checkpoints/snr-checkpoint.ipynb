```python
%%cpp

#include <string>
#include <vector>

int numDistances = 5;
int nBinD = 75;
float factor, snr;
//double hrss;
const char* hName;
std::string histname;
std::string d[5] = { "20000", "10000", "5000", "2500", "1250" };

mdc->SetBranchAddress( "factor", &factor );
//mdc->SetBranchAddress(   "hrss", &hrss   );
mdc->SetBranchAddress(    "snr", &snr    );

std::vector <TH1F*> Vsnr;
Vsnr.reserve( numDistances );
for( int i = 0; i < numDistances; ++i ){
    histname = "distance" + std::to_string(i+1);
    hName = histname.c_str();
    Vsnr.push_back(new TH1F( hName, hName, nBinD, 0, 550 ));
}

int entr = mdc->GetEntries();
for( int i = 0; i < entr; ++i ){
    mdc->GetEntry( i );
    for( int j = 1; j < 6; ++j )
        if( factor == j )
            Vsnr[j-1]->Fill( snr );
}

TCanvas  *ct9 = new TCanvas( "ct9", "ct9", 1920, 1080 );
ct9->cd();
ct9->SetGrid();
for( int i = 1; i < 6; ++i ){
    Vsnr[i-1]->Draw("SAME");
    Vsnr[i-1]->SetLineWidth(4);
}

Vsnr[0]->SetLineColor(kBlue);
Vsnr[1]->SetLineColor(kRed);
Vsnr[2]->SetLineColor(kGreen);
Vsnr[3]->SetLineColor(kBlack);
Vsnr[4]->SetLineColor(kViolet);

Vsnr[0]->            SetTitle(  "SNR"   );
Vsnr[0]->GetXaxis()->SetTitle(  "SNR"   );
Vsnr[0]->GetYaxis()->SetTitle( "counts" );

Vsnr[0]->GetXaxis()->CenterTitle();
Vsnr[0]->GetYaxis()->CenterTitle();

TLegend *legend = new TLegend(0.75,0.6,0.925,0.925);
//legend->SetTextFont(132);
//legend->SetTextFont(42);
legend->SetHeader("Distances", "C");


const char* legName;
std:string lName;
for( int i = 0; i < numDistances; ++i ){
    lName   = d[i] + " kPc";
    legName = lName.c_str();
    legend->AddEntry( Vsnr[i], legName, "l" );
}

legend->Draw("SAME");
gStyle->SetGridStyle(7);
gStyle->SetGridWidth(1);
    
ct9->Draw();
ct9->Print(("snrDistribution.pdf"));
```

```python
%%cpp


std::vector <TH1F*> Vhrss;
Vhrss.reserve( numDistances );
for( int i = 0; i < numDistances; ++i ){
    histname = "distance" + std::to_string(i+1);
    hName = histname.c_str();
    Vhrss.push_back(new TH1F( hName, hName, nBinD, 0, 2.4e-21 ));
}

/*
for( int i = 0; i < entr; ++i ){
    mdc->GetEntry(i);
    std::cout << hrss << std::endl;
}*/

double hrss;
mdc->SetBranchAddress(   "hrss", &hrss   );

for( int i = 0; i < entr; ++i ){
    mdc->GetEntry( i );
    for( int j = 1; j < 6; ++j )
        if( factor == j )
            Vhrss[j-1]->Fill( hrss );
}

TCanvas  *ct8 = new TCanvas( "ct8", "ct8", 1920, 1080 );
ct8->cd();
ct8->SetGrid();

for( int i = 1; i < 6; ++i ){
    Vhrss[i-1]->Draw("SAME");
    Vhrss[i-1]->SetLineWidth(4);
}

Vhrss[0]->SetLineColor(kBlue);
Vhrss[1]->SetLineColor(kRed);
Vhrss[2]->SetLineColor(kGreen);
Vhrss[3]->SetLineColor(kBlack);
Vhrss[4]->SetLineColor(kViolet);

Vhrss[0]->            SetTitle(  "hrss"  );
Vhrss[0]->GetXaxis()->SetTitle(  "hrss"  );
Vhrss[0]->GetYaxis()->SetTitle( "counts" );

Vhrss[0]->GetXaxis()->CenterTitle();
Vhrss[0]->GetYaxis()->CenterTitle();

TLegend *legend1 = new TLegend(0.75,0.6,0.925,0.925);
legend1->SetHeader("Distances", "C");

const char* legName1;
std:string lName1;
for( int i = 0; i < numDistances; ++i ){
    lName1   = d[i] + " kPc";
    legName1 = lName1.c_str();
    legend1->AddEntry( Vhrss[i], legName1, "l" );
}

legend1->Draw("SAME");
gStyle->SetGridStyle(7);
gStyle->SetGridWidth(1);
    
ct8->Draw();
ct8->Print(("hrssDistribution.pdf"));
```
